```python
import sys
import socket
import selectors
import threading
import eel

HOST = '192.168.1.8'
PORT = 9999
statu_server=False


sock_list=[]
class sock_t:
    def __init__(self,a):
        self.statu=a
    pass
sock_statu=[]

with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as sock:
    sock.bind((HOST, PORT))
    sock.listen(5)
    print("Server is listening at {}:{}".format(HOST, PORT))
    sock.setblocking(False)
    with selectors.DefaultSelector() as selector:
        selector.register(sock, selectors.EVENT_READ)
        try:
            while True:
                if statu_server:
                    break
                ready = selector.select(timeout=10)
                if not ready:
                    print("timeout")
                    continue
                for key, _ in ready:
                    if key.fileobj == sock:
                        try:
                            connection, client_address = sock.accept()
                            print("Accepted from", client_address)
                        except:
                            sys.exc_info()[1]
                        else:
                            selector.register(connection, selectors.EVENT_READ)
                            sock_list.append(connection)
                            temp=sock_t(True)
                            sock_statu.append(temp)
                    else:
                        connection = key.fileobj
                        raw_data = connection.recv(1024)
                        if not raw_data:
                            print("Closed connection to", connection.getpeername())
                            selector.unregister(connection)
                            temp=sock_list.index(connection)
                            sock_list.pop(temp)
                            sock_statu.pop(temp)
                            connection.close()
                            break
                        print("Received {} from {}".format(raw_data.decode(), connection.getpeername()))
        except KeyboardInterrupt:
            print("EXIT")
        except:
            sys.exc_info()[1]
            
```
